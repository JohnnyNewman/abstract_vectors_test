```python
from gensim.models.word2vec import *
model = Word2Vec.load("vectors.txt")
```

```python
from pprint import pprint  # pretty-printer
pprint(model.wv.most_similar_cosmul(positive=["evolutionary_algorithm"], topn=10))
```

```python
input_list = [
    ["genetic_algorithm"], [["algorithm", "evolutionary"]], 
    #[["genetic_algorithm","simulated_annealing", "particle_swarm"],["metaheuristic"]],
    [["multi_objective", "optimization"]],
    ["automobile"], [["automobile", "safety"]],  [["vehicle_body", "wing"]],
    ["vehicle_body"], ["chassis"], ["crash"], ["crashworthiness"],
    [["research", "theory"], ["real_world", "industry"]],
    [["real_world", "industry"], ["research", "theory"]],
    ["cooling"], ["heat_transfer"], [["heat_transfer"], ["ice"]], 
    ["topology_optimization"], ["level_set"], [["boundary_condition"]],
    ["shape"], ["sensitivity_analysis"], ["sensitivity"], ["adjoint"], ["gradient"]
             ]

for input in input_list:
    if len(input) == 1:
        print "positive:", input[0]
        pprint(model.wv.most_similar(positive=input[0], topn=20))
    elif len(input) == 2:
        print "positive:", input[0]
        print "negative:", input[1]
        pprint(model.wv.most_similar(positive=input[0], negative=input[1], topn=20))
    print("\n")

print "done"
```
